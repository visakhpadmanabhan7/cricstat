# EDA for ball by ball data

```python
ball_by_ball = pd.read_csv('./ODI_ball_by_ball_data.csv')
ball_by_ball
```

```python
ball_by_ball.dtypes
```

```python
for i in ball_by_ball.columns:
    print(i,ball_by_ball[i].nunique())
    print()
```

```python
ball_by_ball['id'].nunique()==match_data['id'].nunique()
```

## The id value count is same,so it is like a foreign key to match_data,so can bring the date column to this dataframe

```python
ball_by_ball['date'] = ball_by_ball['id'].map(match_data.set_index('id')['date'])
```

### Added date column to ball_by_ball dataframe

```python
#extract year from ball_by_ball['date']
ball_by_ball['year'] = ball_by_ball['date'].dt.year
match_data['year'] = match_data['date'].dt.year
```

```python
plt.figure(figsize=(10,5))
match_data['year'].value_counts().plot(kind='bar')
plt.xticks(rotation=45)
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Year Distribution')
plt.show()
```

```python
plt.figure(figsize=(10,5))
match_data['team1'].value_counts().plot(kind='bar')
plt.xlabel('Matches played')
plt.ylabel('Team')
plt.title('Game Distribution')
plt.show()
```

```python
plt.figure(figsize=(10,5))
match_data['team2'].value_counts().plot(kind='bar')
plt.xlabel('Matches played')
plt.ylabel('Team')
plt.title('Game Distribution')
plt.show()
```
